```python
def img_generator(url):
  wd.get(url)
  soup = BeautifulSoup(wd.page_source,'html.parser')
  tag =  soup.find('ul',attrs={'style':'scroll-padding-top:0;scroll-behavior:smooth'})
  imgs = []
  for img in tag.find_all('img'):
    temp = img['src']
    img = re.sub(r'imwidth=156','imwidth=800',temp)
    imgs.append(img)

  # If total images less than 5, then we skip this page
  if len(imgs)<5:
    imgs = []
  else:
    # We see that the cloth url has an additional '&filter=packshot' in it's url, thus it we don't have to care about it's position now. We will choose the url having this, and name it as cloth
    # We first initialize the idx of this cloth image to be -1. If there isn't any cloth image in the given set of urls we simply skip this page_url and move on to the next one
    idx = -1
    for i,img in enumerate(imgs):
      if re.search('filter=packshot',img):
        idx = i
        # print('this is the cloth url  ', img,'\n', 'and it was find at index ', idx)
    if idx==-1|idx==0|idx==1:
      imgs=[]
  return imgs  
```

```python
def save( s , img_category,img_urls,description):
  global count
  if len(img_urls)<1 :
    print(description , ' this product was returned')
    return




  if img_urls[0]=='':
    print(description , ' this product was returned')
    return
  if img_urls[1]=='':
    print(description , ' this product was returned')
    return


  file_name = img_category + str(count) + '_0'
  with open('0/' + file_name + '.jpg','wb') as f:
    try:
      im = s.get(img_urls[0], headers=HEADERS)
    except ConnectionError as err:
      time.sleep(10) 
      im = s.get(img_urls[0], headers=HEADERS)
    f.write(im.content)   

  file_name = img_category + str(count) + '_1'
  with open('1/' + file_name + '.jpg','wb') as f:
    try:
      im = s.get(img_urls[1], headers=HEADERS)
    except ConnectionError as err:
      time.sleep(10) 
      im = s.get(img_urls[1], headers=HEADERS)
    f.write(im.content) 

  file_name = img_category + str(count) + '_2'
  with open('2/' + file_name + '.jpg','wb') as f:
    try:
      im = s.get(img_urls[2], headers=HEADERS)
    except ConnectionError as err:
      time.sleep(10) 
      im = s.get(img_urls[2], headers=HEADERS)
    f.write(im.content) 

  print(img_category + str(count) + '    name is  ' + description)
  count=count+1
```

```python
def imgs_generator(wd, url):
  wd.get(url)
  soup = BeautifulSoup(wd.page_source,'html.parser')
  tag =  soup.find('ul',attrs={'style':'scroll-padding-top:0;scroll-behavior:smooth'})
  imgs = []
  for img in tag.find_all('img'):
    temp = img['src']
    img = re.sub(r'imwidth=156','imwidth=800',temp)
    imgs.append(img)

  # If total images less than 5, then we skip this page
  if len(imgs)<5:
    imgs = []
  else:
    # We see that the cloth url has an additional '&filter=packshot' in it's url, thus it we don't have to care about it's position now. We will choose the url having this, and name it as cloth
    # We first initialize the idx of this cloth image to be -1. If there isn't any cloth image in the given set of urls we simply skip this page_url and move on to the next one
    idx = -1
    for i,img in enumerate(imgs):
      if re.search('filter=packshot',img):
        idx = i
        # print('this is the cloth url  ', img,'\n', 'and it was find at index ', idx)
    if idx==-1|idx==0|idx==1:
      imgs=[]
    else:
      copy=imgs
      imgs = [copy[idx],copy[1],copy[0]]
  return imgs 
```

```python
def scrape(wd, session, url , ctg):
  print('urls_for_this_page_are_being_generated')
  wd.get(url)
  wd.implicitly_wait(1)
  print(wd.current_url)
  if wd.current_url!= url:
    print('Something fishy, please check the url')
    return 0

  soup = BeautifulSoup(wd.page_source,'html.parser')
  if len(soup.find_all('article',attrs={'role':'link'}))<10:
    wd.close()
    wd = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
    wd.get(url)
    wd.implicitly_wait(1)
    soup = BeautifulSoup(wd.page_source,'html.parser')
  urls = []
  if soup.find_all('article',attrs={'role':'link'}) is not None:
    for tag in soup.find_all('article',attrs={'role':'link'}):
      if re.search(r'https://en.zalando.de',tag.find('a')['href']):
        url = tag.find('a')['href']
        # print(url)
        urls.append(url)
      else:
        url =  'https://en.zalando.de'  + tag.find('a')['href']
        # print(url)
        urls.append(url)
  else:
    print('For this page cant find its products a hrefs')
    return 1

  print('urls have been generated, now we will start scraping')
  print('total_count_for_this_category_is ' ,len(urls),' but many images dont have either cloth image or model image thus they will be discraded')

  # soup = BeautifulSoup(wd.page_source, 'html.parser')
  # urls = []
  # page_url = url

  # if soup.find_all('article',attrs={'role':'link'}) is not None:
  #   for tag in soup.find_all('article',attrs={'role':'link'}):
  #     if re.search(r'zalando.de',tag.find('a')['href']):
  #       url = tag.find('a')['href']
  #     else:
  #       url =  'https://en.zalando.de'  + tag.find('a')['href']
  #     # print(url)
  #     urls.append(url)
  # else:
  #   print('For this product cant find its a href')
  #   return 1

  # print('urls have been generated, now we will start scraping')
  # print('total_count_for_this_category_is ' ,len(urls),' but many images dont have either cloth image or model image thus they will be discraded')

  for url in urls:
    imgs = []
    wd.get(url)
    wd.implicitly_wait(0.3)
    soup = BeautifulSoup(wd.page_source,'html.parser')
    tag =  soup.find('ul',attrs={'style':'scroll-padding-top:0;scroll-behavior:smooth'})

    if tag is not None:
      if tag.find_all('img') is not None:
        for img in tag.find_all('img'):
          temp = img['src']
          img = re.sub(r'imwidth=156','imwidth=800',temp)
          imgs.append(img)
      else:
        print(func_description(url),' for this product <ul> tag containing imgs found, but imgs are no there ,moving on to the next product')
        continue
    else:
      print(func_description(url) , ' cant find <ul> tag which contains images for this product, moving on to the next product')
      continue










    # If total images less than 5, then we skip this page
    if len(imgs)<5:
      imgs = []
    else:
      # We see that the cloth url has an additional '&filter=packshot' in it's url, thus it we don't have to care about it's position now. We will choose the url having this, and name it as cloth
      # We first initialize the idx of this cloth image to be -1. If there isn't any cloth image in the given set of urls we simply skip this page_url and move on to the next one
      idx = -1
      for i,img in enumerate(imgs):
        if re.search('filter=packshot',img):
          idx = i
          # print('this is the cloth url  ', img,'\n', 'and it was find at index ', idx)
      if idx==-1|idx==0|idx==1:
        imgs=[]
      else:
        copy=imgs
        imgs = [copy[idx],copy[1],copy[0]]
    save( session,  ctg ,       imgs     ,   func_description(url) ) 
  return 1
```
